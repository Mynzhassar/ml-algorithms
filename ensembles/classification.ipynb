```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer = load_breast_cancer()

data = breast_cancer.data
target = breast_cancer.target

data.shape
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
%%time

rf_score = np.mean(cross_val_score(RandomForestRegressor(), data, target))
rf_score
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
%%time

sklearn_gb_score = np.mean(cross_val_score(GradientBoostingClassifier(), data, target))
sklearn_gb_score
```

### XGBoost

```python
from xgboost import XGBClassifier
```

```python
%%time

xgboost_score = np.mean(cross_val_score(XGBClassifier(), data, target))
xgboost_score
```

### LightGBM

```python
from lightgbm import LGBMClassifier
```

```python
%%time

lightgbm_score = np.mean(cross_val_score(LGBMClassifier(), data, target))
lightgbm_score
```

### CatBoost

```python
from catboost import CatBoostClassifier
```

```python
%%time

catboost_score = np.mean(cross_val_score(CatBoostClassifier(), data, target))
catboost_score
```
