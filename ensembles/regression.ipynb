```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
from sklearn.datasets import load_boston
from sklearn.model_selection import cross_val_score
```

```python
boston = load_boston()

data = boston.data
target = boston.target

data.shape
```

```python
scoring='neg_mean_squared_error'
```

### Random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
%%time

rf_score = np.abs(np.mean(cross_val_score(RandomForestRegressor(), data, target, scoring=scoring)))
rf_score
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
%%time 

gb_score = np.abs(np.mean(cross_val_score(GradientBoostingRegressor(), data, target, scoring=scoring)))
gb_score
```

### XGBoost

```python
from xgboost import XGBRegressor
```

```python
%%time 

xgb_score = np.abs(np.mean(cross_val_score(XGBRegressor(), data, target, scoring=scoring)))
xgb_score
```

### LightGBM

```python
from lightgbm import LGBMRegressor
```

```python
%%time 

lgb_score = np.abs(np.mean(cross_val_score(LGBMRegressor(), data, target, scoring=scoring)))
lgb_score
```

### CatBoost

```python
from catboost import CatBoostRegressor
```

```python
%%time 

cb_score = np.abs(np.mean(cross_val_score(CatBoostRegressor(), data, target, scoring=scoring)))
cb_score
```
